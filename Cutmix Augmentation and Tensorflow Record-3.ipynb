# Introduction to Data Augmentation and TensorFlow Records

- `RandomFlip`: Randomly flips images horizontally.
- `RandomContrast`: Applies random contrast adjustments to images.

The `box` function calculates the coordinates of a bounding box based on a parameter called `lamda`. The bounding box is defined by its top-left corner coordinates (`r_x` and `r_y`) and its dimensions (`r_h` and `r_w`).

The `fit_label` function simply converts a label to the `tf.float32` data type, which can be useful for machine learning tasks.

## CutMix Data Augmentation

The `cutmix` function applies the CutMix data augmentation technique. It combines parts of two input images, `image_1` and `image_2`, to create a new image, and it calculates a new label based on the combination.

## Dataset Augmentation

Finally, the code loads and parses the TensorFlow Records into a dataset called `parsed_data`, resizing images and one-hot encoding labels to prepare the data for training.

## Explanation of the `box` Function

The `box` function is a crucial component in image data manipulation, particularly in the context of data augmentation for machine learning. It calculates the coordinates and dimensions of a bounding box based on the input parameter `lamda`. Let's break down the key aspects of this function:

1. **Random Top-Left Corner Coordinates (`r_x` and `r_y`)**:
   - The `box` function starts by generating two random values, `r_x` and `r_y`, using a uniform distribution with a range between 0 and `img_size`.
   - These values represent potential top-left corner coordinates of the bounding box.
   - `tf.cast` is used to ensure that these values are integers of data type `tf.int32`.

2. **Bounding Box Dimensions (`r_w` and `r_h`)**:
   - The dimensions of the bounding box, represented as `r_w` (width) and `r_h` (height), are computed based on the square root of `(1 - lamda)` times `img_size`.
   - `tf.cast` is applied to maintain these dimensions as integers of data type `tf.int32`.

3. **Bounding Box Clipping**:
   - To ensure that the bounding box remains within the image boundaries, the code adjusts the top-left coordinates `r_x` and `r_y` using `tf.clip_by_value`. This ensures that the values do not fall below 0 or exceed `img_size`.

4. **Calculation of Bottom-Right Corner Coordinates (`x_b_r` and `y_b_r`)**:
   - `x_b_r` and `y_b_r` are calculated by adding the width and height to the top-left coordinates (`r_x` and `r_y`, respectively).
   - These values are also clipped using `tf.clip_by_value` to ensure they remain within the image boundaries.

5. **Bounding Box Dimensions Adjustment**:
   - The code checks if the width (`r_w`) or height (`r_h`) of the bounding box is equal to 0. If so, they are adjusted to a minimum value of 1 to prevent the creation of zero-sized bounding boxes.

6. **Return Values**:
   - The function returns a tuple consisting of the top-left corner coordinates (`r_y`, `r_x`) and the dimensions of the bounding box (`r_h`, `r_w`).

The `box` function is a fundamental tool for image data augmentation and manipulation in machine learning. It is responsible for generating the necessary parameters to define a bounding box that remains within image boundaries and has appropriate dimensions. This bounding box is commonly used for tasks like cropping, resizing, and various data augmentation techniques to enhance training data for machine learning models.


```python
def box(lamda):

    r_x = tf.cast(tfp.distributions.Uniform(low=0, high=img_size).sample(1)[0], dtype=tf.int32)
    r_y = tf.cast(tfp.distributions.Uniform(low=0, high=img_size).sample(1)[0], dtype=tf.int32)

    r_w = tf.cast(img_size * tf.math.sqrt(1-lamda), dtype=tf.int32)
    r_h = tf.cast(img_size * tf.math.sqrt(1-lamda), dtype=tf.int32)

    r_x = tf.clip_by_value(r_x - r_w//2, 0, img_size)
    r_y = tf.clip_by_value(r_y - r_h//2, 0, img_size)

    x_b_r = tf.clip_by_value(r_x + r_w//2, 0, img_size)
    y_b_r = tf.clip_by_value(r_y + r_h//2, 0, img_size)

    r_w = x_b_r - r_x
    if r_w == 0:
        r_w = 1

    r_h = y_b_r - r_y
    if r_h == 0:
        r_h = 1

    return r_y, r_x, r_h, r_w
```

## Explanation of the `cutmix` Function

The `cutmix` function is a crucial component of image data augmentation and manipulation, primarily used for enhancing training data in machine learning. Let's dissect the key elements of this function:

1. **Input Parameters**:
   - The function takes two input datasets, `train_dataset_1` and `train_dataset_2`. Each dataset consists of pairs of images and their corresponding labels, represented as `(image, label)`.

2. **Beta Distribution for `lamda`**:
   - The variable `lamda` is generated using a Beta distribution with parameters `(0.2, 0.2)`. This value `lamda` serves as a mixing coefficient that determines the extent of blending between the two images.
   - A random value is sampled from this distribution to create a blend ratio.

3. **Bounding Box Calculation**:
   - The function calculates the coordinates and dimensions of a bounding box using the `box` function, which depends on the `lamda` value.
   - The bounding box's top-left corner coordinates are determined by `r_x` and `r_y`, and its dimensions are specified as `r_h` (height) and `r_w` (width).

4. **Image Cropping and Padding**:
   - Using the calculated bounding box, the function performs cropping and padding operations on both input images, `image_1` and `image_2`.
   - It crops a region from each image based on the bounding box's specifications and pads it to match the original image size (`img_size`).

5. **Image Blending**:
   - To create the final mixed image, the function blends the two cropped and padded images, `pad_1` and `pad_2`, using the `image_1`.
   - This blending is achieved by subtracting `pad_1` from `image_1` and adding `pad_2`.

6. **Label Adjustment**:
   - The function adjusts the label based on the mixing coefficient `lamda` and the bounding box's dimensions.
   - The adjusted label is a linear combination of the labels from `label_1` and `label_2`, where the weights are determined by `lamda` and `1 - lamda`.

7. **Return Values**:
   - The function returns the mixed image and the adjusted label.

The `cutmix` function plays a pivotal role in generating augmented training data for machine learning models. It combines two images with a controlled blending ratio and adjusts the labels accordingly, contributing to the diversification and enrichment of the training dataset.


```python
img_size = configrations["image_size"]
def cutmix(train_dataset_1, train_dataset_2):

    (image_1, label_1), (image_2, label_2) = train_dataset_1, train_dataset_2


    lamda = tfp.distributions.Beta(0.2, 0.2)
    lamda = lamda.sample(1)[0]

    r_y, r_x, r_h, r_w = box(lamda)

    crop_1 = tf.image.crop_to_bounding_box(
        image=image_1, offset_height=r_y, offset_width=r_x, target_height=r_h, target_width=r_w)
    pad_1 = tf.image.pad_to_bounding_box(crop_1, r_y, r_x, img_size, img_size)

    crop_2 = tf.image.crop_to_bounding_box(
        image=image_2, offset_height=r_y, offset_width=r_x, target_height=r_h, target_width=r_w)
    pad_2 = tf.image.pad_to_bounding_box(crop_2, r_y, r_x, img_size, img_size)

    image = image_1 - pad_1 + pad_2

    lamda = fit_label(1 - (r_w * r_h) / (img_size * img_size))

    label = lamda*fit_label(label_1) + (1-lamda) * fit_label(label_2)


    return image, label
```

## Preparing Data for TensorFlow Records (TFRecords)

TFRecords are a common data format used for efficient and compact storage of large datasets in TensorFlow. The code snippet illustrates the process of creating and saving data in the TFRecords format:

1. **`create_example` Function**:
   - The `create_example` function is defined to convert an image and its corresponding label into a serialized TFRecord example.
   - It uses the `Example` class to construct an example with two features: 'images' and 'labels'.
   - The 'images' feature is encoded as bytes, and the 'labels' feature is encoded as an integer.
   - The function returns the serialized TFRecord example using `SerializeToString()`.

2. **Creating Directories**:
   - Two directories are created using the `! mkdir` shell command to store the TFRecords. The directories are named "/content/tensorflow record" and "/content/tensorflow record/tfrecords/".

3. **Setting the Number of Shards**:
   - The variable `num_shards` is set to 10. This represents the number of TFRecord files to be created to store the data. Sharding is often used to distribute the data across multiple files for parallel processing.

4. **Defining the File Path Template**:
   - The `example_path` variable is defined as the template for the file paths of the TFRecord files. It contains a placeholder for the shard number.

The code provides the infrastructure for saving data in the TFRecords format, a standard practice for efficiently handling and loading large datasets in TensorFlow. The data is serialized into a format that can be easily read and processed for machine learning tasks.


```python
def create_example(image, label):

    bytes_feature = Feature(
        bytes_list = BytesList(value=[image]))

    int_feature = Feature(
        int64_list = Int64List(value=[label]))

    example = Example(
        features = Features(feature={
            'images': bytes_feature,
            'labels': int_feature,
        }))

    return example.SerializeToString()
```

## Encoding Images and Labels for TFRecords

This code snippet demonstrates the process of encoding images and labels in preparation for saving them as TFRecords. The encoding involves converting images to JPEG format and mapping the function to the training dataset:

1. **`encode_image` Function**:
   - The `encode_image` function takes an image and its corresponding label as input.
   - The image is converted to an 8-bit unsigned integer format using `tf.image.convert_image_dtype`. This step is necessary to ensure consistent encoding.
   - The image is then encoded as a JPEG image using `tf.io.encode_jpeg`.
   - The function returns the encoded image and the label's index corresponding to the class.

2. **Mapping the Function**:
   - The `encode_image` function is mapped to the `train_data` dataset using the `.map` method. This process applies the image and label encoding to each example in the dataset.

3. **Encoded Dataset**:
   - The result of the mapping is stored in the `encoded_dataset` variable. This dataset contains pairs of encoded images and label indices.

The code prepares the data by converting images to a common format (JPEG) and encoding labels as integer indices, making them suitable for storage as TFRecords. This encoding step is crucial for efficient data storage and retrieval in TensorFlow.


```python
def encode_image(image, label):
    image = tf.image.convert_image_dtype(image, dtype=tf.uint8)
    image = tf.io.encode_jpeg(image)
    return image, tf.argmax(label)
```

## Writing Encoded Data to TFRecord Files and Creating a Dataset

This code snippet demonstrates the process of writing the encoded data to TFRecord files and creating a dataset from those files. This is a common practice in machine learning for efficient data storage and retrieval:

1. **Sharding Data**:
   - A loop iterates through a range of `shard_num` values from 0 to `num_shards` - 1.
   - The encoded data from `encoded_dataset` is sharded into `num_shards` segments, each containing a portion of the data.
   - `sharded_data` is created as an iterator containing the sharded data.

2. **Writing TFRecord Files**:
   - Within the loop, a TFRecord file is created for each shard using the `tf.io.TFRecordWriter`. The file path is determined by the `example_path` template with the `shard_num` placeholder.
   - For each shard, the sharded data is iterated through using `for encoded_image, encoded_iabel in sharded_data:`.
   - An example is created for each pair of encoded image and label using the `create_example` function.
   - The example is written to the TFRecord file using `file_writer.write(example)`.

3. **Generating File Paths for Reading**:
   - A list `l` is created, containing file paths for all the written TFRecord files. The file paths are generated by filling in the shard number placeholder in the `example_path` template.

4. **Creating a Dataset from TFRecord Files**:
   - A `tf.data.TFRecordDataset` is created using the file paths in `l`. This dataset is used for reading the data from the TFRecord files.

The code completes the process of writing encoded data to TFRecord files, making the data ready for efficient storage and retrieval. Additionally, it sets up a dataset to read and use the TFRecord data in machine learning workflows.


```python
for shard_num in range(num_shards):

    sharded_data = (
        encoded_dataset
        .shard(num_shards, shard_num)
        .as_numpy_iterator()
    )

    with tf.io.TFRecordWriter(example_path.format(shard_num)) as file_writer:
        for encoded_image, encoded_iabel in sharded_data:
            example = create_example(encoded_image, encoded_iabel)
            file_writer.write(example)
```

```python
l = [example_path.format(p) for p in range(shard_num+1)]

recond_dataset = tf.data.TFRecordDataset(filenames=l)
```

## Parsing TFRecords, Decoding Images, and Creating the Final Dataset

This code snippet outlines the process of parsing TFRecords, decoding image data, and creating the final dataset for machine learning:

1. **`parse_tfrecords` Function**:
   - The `parse_tfrecords` function is defined to parse individual TFRecord examples.
   - It specifies the structure of the TFRecord data using `feature_description`, which defines the expected features and their data types.
   - The function parses a single TFRecord example using `tf.io.parse_single_example`, matching the features based on the description.
   - It decodes the 'images' feature from JPEG format to a 3-channel tensor using `tf.io.decode_jpeg`.

2. **Mapping and Batching Data**:
   - The `recond_dataset`, which contains TFRecord data, is mapped to the `parse_tfrecords` function using `.map`. This step parses and decodes the TFRecord data.
   - The parsed data is then batched into batches of a specified size (configurations["batch_size"]) using `.batch`.
   - Batching is important for training machine learning models in mini-batches, which helps in memory management and optimization.

3. **Prefetching (Optional)**:
   - The `.prefetch` method is used to prefetch data from the dataset for better performance. It enables the preparation of the next batch while the current batch is being processed, reducing training time. The value `tf.data.AUTOTUNE` optimizes the prefetching process.

The code snippet demonstrates the crucial steps in parsing, decoding, and batching TFRecord data to create the final dataset. This dataset is ready for use in machine learning tasks, such as training deep learning models.


```python
def parse_tfrecords(example):

    feature_description = {
        "images": tf.io.FixedLenFeature([], tf.string),
        "labels": tf.io.FixedLenFeature([], tf.int64)
    }

    example = tf.io.parse_single_example(example, feature_description)
    example["images"] = tf.io.decode_jpeg(example["images"], channels=3)

    return example["images"], example["labels"]
```

```python
parsed_data = (
    recond_dataset
    .map(parse_tfrecords)
    .batch(configrations["batch_size"])
    .prefetch(tf.data.AUTOTUNE)
)
```

## Training a Neural Network

In this code snippet, a neural network is being trained to perform a specific task, likely an image classification task. The training process involves the following key components:

1. **Model Checkpointing (`ModelCheckpoint`)**:
   - The `ModelCheckpoint` callback is set up to monitor the validation accuracy (`val_accuracy`).
   - It saves the model with the best validation accuracy as "Emotion_model.h5" in case the validation accuracy improves. The `save_best_only` parameter ensures that only the best model is saved.
   - The callback is used to prevent the loss of the best model if the training process is interrupted.

2. **Learning Rate Reduction (`ReduceLROnPlateau`)**:
   - The `ReduceLROnPlateau` callback monitors the validation accuracy and adjusts the learning rate (`lr`) during training.
   - If the validation accuracy does not improve for a certain number of epochs (specified by `patience`), the learning rate is reduced by a factor. In this case, the minimum learning rate is set to 1e-5.
   - This is a common technique to fine-tune learning rates for better convergence.

3. **Early Stopping (`EarlyStopping`)**:
   - The `EarlyStopping` callback monitors the validation accuracy and stops training if the validation accuracy does not improve for a certain number of epochs (specified by `patience`).
   - This prevents overfitting and saves training time.

4. **Training Loop**:
   - The model is trained using the `model.fit` function, which takes the training data and validation data as inputs.
   - The training process runs for a specified number of epochs (as defined in `configurations["n_epochs"]`).
   - Callbacks, including model checkpointing, learning rate reduction, and early stopping, are passed to the `callbacks` parameter of `model.fit`.

5. **Training Progress**:
   - During training, the code prints information for each epoch, including the loss, accuracy, and learning rate. It also mentions whether the validation accuracy improved or if the learning rate was reduced.
   - The training process involves multiple epochs, and the training may stop early if the validation accuracy does not improve.

```python
check_point = ModelCheckpoint(
    "/content/Emotion_model.h5",
    monitor="val_accuracy",
    save_best_only=True,
    verbose=1
)

redlr = ReduceLROnPlateau(
    monitor='val_accuracy',
    patience=3,
    min_lr=1e-5,
    verbose=1
)

es = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    verbose=1
)
```
